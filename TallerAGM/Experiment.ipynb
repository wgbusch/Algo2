```python
from src.python.segment_image import segment_image, read_image, rgb2gray
import matplotlib.pyplot as plt
```

```python
def show_results(file_name, k_to_try):
    original = plt.imread(file_name)
    segmentations = [segment_image(file_name, k=k) for k in k_to_try]

    fig, axs = plt.subplots(1, len(k_to_try) + 1, figsize=(15, 3))
    axs[0].imshow(original)
    axs[0]
    i = 0
    for ax in axs[1:]:
        ax.imshow(segmentations[i])
        i = i + 1
```

```python
show_results('images/1.jpg', [1, 20000])
```
